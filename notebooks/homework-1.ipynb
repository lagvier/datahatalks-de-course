```python
import pandas as pd
```

```python
df = pd.read_csv("green_tripdata_2019-09.csv.gz")
df.head()
```

```python
df.columns
```

```python
bd = pd.read_csv('taxi+_zone_lookup.csv')
bd.head()
```

```python
df.lpep_pickup_datetime=pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime=pd.to_datetime(df.lpep_dropoff_datetime)
df['pd'] = df['lpep_pickup_datetime'].dt.date.astype('str')
df['dd'] = df['lpep_dropoff_datetime'].dt.date.astype('str')
df['time'] = (df.lpep_dropoff_datetime - df.lpep_pickup_datetime)
```

```python
df1 = df[(df['pd'] == '2019-09-18') & (df['dd']=='2019-09-18')]
df1.shape
```

```python
df.sort_values(by='time', ascending=False).head()
```

```python
df2 = df[(df['pd'] == '2019-09-18')]
df2['LocationID'] = df2['PULocationID']
df2 = df2.merge(bd, on='LocationID', how='left')
df2.shape, df2.groupby('Borough')['total_amount'].sum()
```

```python
df2 = df[["PULocationID","DOLocationID","tip_amount"]]
bd1 = bd[["LocationID","Zone"]].rename(columns={"LocationID": "PULocationID"})
df2 = df2.merge(bd1, on='PULocationID', how='left')
df2 = df2.rename(columns={"Zone": "PUZone"})

bd1 = bd1.rename(columns={"PULocationID": "DOLocationID"})
df2 = df2.merge(bd1, on='DOLocationID', how='left')
df2 = df2.rename(columns={"Zone": "DOZone"})
df2[df2['PUZone']=='Astoria'].head()
```
